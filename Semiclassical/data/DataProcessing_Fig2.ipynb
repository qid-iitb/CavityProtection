```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
from matplotlib import *

from matplotlib import rc
plt.rc('font',**{'family':'serif','sans-serif':['Computer Modern Sans serif'],'size':16})
plt.rcParams['text.usetex'] = True

#rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
# font.family        : serif
# font.serif         : Times, Palatino, New Century Schoolbook, Bookman, Computer Modern Roman
# font.sans-serif    : Helvetica, Avant Garde, Computer Modern Sans serif
# font.cursive       : Zapf Chancery
# font.monospace     : Courier, Computer Modern Typewriter
# text.usetex        : true
```

```python
data1 = np.genfromtxt('r0Osc.dat')
data2 = np.genfromtxt('r1Osc.dat')
data3 = np.genfromtxt('r2Osc.dat')

data4 = np.transpose(np.genfromtxt('Data_d60.dat'))
data5 = np.transpose(np.genfromtxt('Data_d70.dat'))
data6 = np.transpose(np.genfromtxt('Data_d80.dat'))
```

```python
fig1, (ax0,ax1) = plt.subplots(1,2,gridspec_kw={'wspace':0.25},figsize=(10,3.6))

ax0.set_xlabel(r'Time (in units of $2\pi/\Omega$)',labelpad=6,fontsize=18)
ax0.set_ylabel(r'Photon number $n(t)$',labelpad=6,fontsize=18)

ax1.set_xlabel(r'Squeezing parameter $r$',labelpad=6,fontsize=18)
ax1.set_ylabel(r'$\Gamma_n$ (in units of ${\Omega}/2\pi$)',labelpad=4,fontsize=18)

cmap = plt.get_cmap('rainbow')
color = ['red', 'blue', 'black']

#Divided time axis by 2*Pi/4 to scale time in proper units of collective coupling strength
ax0.plot(data1[:,0]*4/(2*np.pi),data1[:,-1],color=color[0],label=r"$r=0.0$",linestyle='--') 
ax0.plot(data2[:,0]*4/(2*np.pi),data2[:,-1],color=color[1],label=r"$r=1.0$",linestyle='-.')
ax0.plot(data3[:,0]*4/(2*np.pi),data3[:,-1],color=color[2],label=r"$r=2.0$",linestyle=':')

ax0.set_xlim(0,1)
ax0.set_ylim(0,1)
ax0.text(-0.22, 1.042, "(a)", va='center',fontsize=18)
ax1.text(1.125, 14.5, "(b)", va='center',fontsize=18)

ax1.errorbar(data4[0],data4[1]*2*np.pi/4,yerr=data4[2]*2*np.pi/4,color=color[0],label=r"$\delta=60$ MHz",marker='o',linestyle=':')
ax1.errorbar(data5[0],data5[1]*2*np.pi/4,yerr=data5[2]*2*np.pi/4,color=color[1],label=r"$\delta=70$ MHz",marker='x',linestyle=':')
ax1.errorbar(data6[0],data6[1]*2*np.pi/4,yerr=data6[2]*2*np.pi/4,color=color[2],label=r"$\delta=80$ MHz",marker='s',linestyle=':')
ax1.set_xlim(1.35,2.45)
ax1.set_ylim(0.0,14)
ax0.tick_params(axis='both', which='major', labelsize=16)
ax1.tick_params(axis='both', which='major', labelsize=16)
ax1.set_xticks([1.4,1.6,1.8,2.0,2.2,2.4])
ax1.set_yticks([0,2,4,6,8,10,12,14])

plt.savefig('Fig2.pdf',bbox_inches='tight',dpi=300)
plt.show()
```
